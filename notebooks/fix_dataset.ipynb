```python
import pandas as pd
```

```python
dataset = pd.read_csv('../datasets/new_DINO_gold_dataset_with_prompt_boxes.csv')
```

```python
dataset
```

```python
def fix_prompt(promt):
    return f"[INST] {promt} [/INST]"

def add_pad(answer):
    return f"{answer} </s>"
```

```python
dataset['correct'] = dataset['correct'].apply(add_pad)
```

```python
dataset['rejected'] = dataset['rejected'].apply(add_pad)
```

```python
p = 0.7

print("Correct!")
dataset['correct_score_2'] = 0
dataset['correct_score_2'][dataset['iou_correct']>p] = dataset[dataset['iou_correct']>p]['score_correct']
score_correct = dataset['correct_score_2'].mean()
print(score_correct)
print()
print("Rejected!")
dataset['rejected_score_2'] = 0
dataset['rejected_score_2'][dataset['iou_rejected']>p] = dataset[dataset['iou_rejected']>p]['score_rejected']
score_rejected = dataset['rejected_score_2'].mean()
print(score_rejected)
print()
print("Margin!")
print(score_correct - score_rejected)
```
